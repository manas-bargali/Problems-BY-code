```python
class Solution:
    def BinarySearch(self, l, key, n):
        # assume ascending ordered list
        low = 0
        high = n - 1
        while low <= high:
            mid = (low + high) // 2
            if l[mid] == key:
                return mid
            elif l[mid] < key:
                low = mid + 1
            else:
                high = mid - 1
        return -1
```

```python
obj = Solution()


result1 = obj.BinarySearch([1.3, 5.6, 7.201], 7, len([1.3, 5.6, 7.201]))
print(result1) 


result2 = obj.BinarySearch([1, 3.5, 6.7, 20.7], 1, len([1, 3.5, 6.7, 20.7]))
print(result2) 
```
