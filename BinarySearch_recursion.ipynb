```python
class Solution:
    def BinarySearch(self, l, low, high, key):
        if low == high:
            if l[low] == key:
                return low
            else:
                return -1
        else:
            mid = (low + high) // 2
            if l[mid] == key:
                return mid
            elif l[mid] < key:
                return self.BinarySearch(l, mid + 1, high, key)
            else:
                return self.BinarySearch(l, low, mid - 1, key)
```

```python

obj = Solution()
result = obj.BinarySearch([1, 2, 3, 4, 5, 8, 10], 0, 6, 8)
print(result)  
```
